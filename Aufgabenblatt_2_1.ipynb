# Aufgabe 1 â€” Erste Schritte mit NumPy und matplotlib


```python
import matplotlib.pyplot as plt
from skimage.io import imread
import numpy as np
img = imread('./catG.png')
plt.imshow(img , cmap='Greys_r', vmin=0, vmax=255)
```

```python
#shape[0] = Zeilen
#shape[1] = Spalten
def countPixels():
  pixelCount = 0
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      pixelCount += 1
  return pixelCount

countPixels()
```

```python
def maxPixelValue():
  maximumValue = 0
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      if(img[x,y] > maximumValue):
        maximumValue = img[x,y]
  return maximumValue

maxPixelValue()
```

```python
def minPixelValue():
  minimumValue = 255
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      if(img[x,y] < minimumValue):
        minimumValue = img[x,y]
  return minimumValue

minPixelValue()
```

```python
def meanPixelValue():
  pixelCount = 0
  pixelSum = 0
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      pixelCount += 1
      pixelSum += img[x,y]

  pixelMean = pixelSum/pixelCount
  return pixelMean

meanPixelValue()
```

# Aufgabe 2 â€” Komplexere Bildstatistiken

```python
def countWhitePixels():
  whitePixelSum = 0
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      if img[x,y] == 255:
        whitePixelSum += 1

  return whitePixelSum

countWhitePixels()
```

```python
def countBlackPixels():
  blackPixelSum = 0
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      if img[x,y] == 0:
        blackPixelSum += 1

  return blackPixelSum

countBlackPixels()
```

```python
def countGreyPixels():
  greyPixelList = []
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      uniqueGreyPixel = img[x,y]
      greyPixelList.append(uniqueGreyPixel)

  for i in range(0,256):
    print(str(i)+": "+str(greyPixelList.count(i)))


countGreyPixels()
```

# Aufgabe 4 â€” Zusatzaufgabe: Einfache Bildtransformationen

```python
def invertImg():
  for x in range(img.shape[0]):
    for y in range(img.shape[1]):
      img[x,y] = 255-img[x,y]

  plt.imshow(img , cmap='Greys_r', vmin=0, vmax=255)

invertImg()
```
